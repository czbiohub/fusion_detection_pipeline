```python
import pandas as pd
import os
```

```python
f = 'cellsList.txt'
s3_path = 's3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_fastqs_9.27/STAR-fus_out/'
! aws s3 ls $s3_path > $f
! cat $f
```

```python
cellsList = pd.read_table('cellsList.txt', delim_whitespace=True, names=['prefix', 'cellName'])
cellsList
```

```python
cellsList = cellsList['cellName']
cellsList
```

```python
cellsList = cellsList.str.strip('/')
cellsList
```

```python
cells_df = pd.DataFrame(data=cellsList, columns=['cellName', 'fullPath'])
cells_df
```

```python
cells_df['fullPath'] = s3_path + cells_df['cellName'] + '/'
cells_df
```

```python
cells_df['fullPath'][10]
```

```python
cells_df['fusionsFile_s3'] = cells_df['fullPath'] + 'star-fusion.fusion_predictions.abridged.tsv'
cells_df
```

```python
cells_df['fusionsFile_s3'][10]
```

```python
cells_df['fusionFile_local'] = cells_df['cellName'] + '_fusions_predictions.abridged.tsv'
cells_df
```

```python
cells_df['cmd'] = 'aws s3 cp ' + cells_df['fusionsFile_s3'] + ' ' + cells_df['fusionFile_local']
cells_df
```

```python
cells_df['cmd'][10]
```

```python
cmds_list = cells_df['cmd']
cmds_list
```

```python
cmds_list_df = pd.DataFrame(data=cmds_list)
cmds_list_df
```

```python
cmds_list_df.to_csv('s3_copy_cmds_list.csv', index=False, header=False)
```
